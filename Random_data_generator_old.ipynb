```python
id_no=1
for i in range(0, len(gender)):
    choice = stream.append(np.random.choice(('IT','COMPS','EXTC', 'MECH'), p=(0.25,0.25,0.25,0.25)))
    year = "2021"
    if choice == 'IT':
        rec = year + "1300"
    if choice == 'COMPS':
        rec = year + "1400"
    if choice == 'EXTC':
        rec = year + "1200"
    else :
        rec = year + "1100"
    if id_no < 10 :
        num = "00" + str(id_no)
    elif id_no < 100 :
        num = "0" + str(id_no)
    else :
        num = str(id_no)
    rec = rec + num
    id_no += 1
    sid.append(rec)
print(stream)
print(sid)
```
