```python
import pandas as pd
import numpy as np
from faker import Faker
```

```python
fake = Faker()
```

```python
name = []
x_perc = []
xii_perc = []
cet_perc = []
physics_xii = []
chem_xii = []
maths_xii = []
jee_perc = []
```

```python
for _ in range(100) :
    name.append(fake.name())
    #generates percentage rounded upto 2 decimals
    x_perc.append(round(np.random.uniform(35.0, 100.0),2))
    xii_perc.append(round(np.random.uniform(51.0, 100.0),2))
    cet_perc.append(round(np.random.uniform(0.0, 100.0),2))    
    physics_xii.append(round(np.random.uniform(50.0, 100.0),2))
    chem_xii.append(round(np.random.uniform(50.0, 100.0),2))
    maths_xii.append(round(np.random.uniform(50.0, 100.0),2))
    jee_perc.append(round(np.random.uniform(0.0, 90.0),2))     
```

```python
data = pd.DataFrame(zip(name,x_perc,xii_perc,cet_perc,physics_xii,chem_xii,maths_xii,jee_perc),columns=['name','x_perc','xii_perc','cet_perc','physics_xii','chem_xii','maths_xii','jee_perc'])
data
```

```python
dataset = data.to_csv('Student_dataset.csv', index = False)
```
